```python
def divide(a, b):
    try:
        c = a/b
        return c
    except ZeroDivisionError:
        print('you can not divide a number by zero')
    except TypeError:
        print('you are not dividing with number enter number and try again')   
```

```python
divide(10, 10)
```

```python
divide(10, 0)
```

```python
divide(10, 'a')
```

```python
def divide(a, b):
    try:
        c = a/b
        return c
    except Exception as e:
      print(e)
```

```python
divide(1, 1)
```

```python
divide(1,0)
```

```python
divide(1,'a')
```

```python
def calculator(operator, a, b):
    if operator == '+':
        return a+b
    elif operator == '-':
        return a-b
    elif operator == '/':
        return a/b
    elif operator == '*':
        return a*b
    elif operator == '**':
        return a**b
    elif operator == '%':
        return a%b

calculator('+', 2, 3)
```

```python
calculator('-',3,2)
```
