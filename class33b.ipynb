```python
import numpy as np
arr2d_a = np.arange(25).reshape(5,5)
arr2d_b = np.arange(25,50).reshape(5,5)
```

```python
print(arr2d_a)
print('----------------------------------')
print(arr2d_b)
```

```python
vertical_joined_array = np.concatenate((arr2d_a,arr2d_b),axis=0)
vertical_joined_array
```

```python
horizontal_joined_array = np.concatenate((arr2d_a,arr2d_b),axis=1)
horizontal_joined_array
```
