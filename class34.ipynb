```python
def updated_email():
    show_email = []
    for name, email in zip(df['Students'], df['Email']):
        email = email.split('@')
        email[0] = name.lower()
        email = '@'.join(email)
        show_email.append(email)
    return show_email
print(len(updated_email()))   
```
