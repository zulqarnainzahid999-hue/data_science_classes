```python
def calculator(op, a, b):
    try:
        return eval(f'a {op} b')
    except Exception as s:
        print(s)
```

```python
calculator('/',5, 'a')
```

```python
def calculator(op, a, b):
    try:
        return eval(f'a {op} b')
    except (ZeroDivisionError, TypeError) as s:
        print(s)
    else:
        print('try run, no exception')
    finally:
        print('it definately run now')
```

```python
calculator('/',5,0)
```

```python
calculator('+', 5,'a')
```

```python
def foo():
    try:
        return 1
    finally:
        return 2
k = foo()
print(k)
```

 Explanation: In Python, if both the try block and the finally block contain return
 statements, the return in the finally block overrides the one in the try block. So,
 even though return 1 is in the try, the function ends up returning 2 because of the
 finally block.

```python
def foo():
    try:
        return (1)
    finally:
        return (2)
k = foo()
print(k)
```

```python
def foo():
    try:
        print(1)
    finally:
        print(2)
foo()
```
