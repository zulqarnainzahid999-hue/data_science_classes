## list
- A list in Python is a built-in, versatile data structure used to store an ordered collection 
of items in a single variable. It is one of Python's most frequently used data types because 
it can hold multiple data types simultaneously, such as integers, strings, and even other lists

```python
nums = [10, 20, 30, 20]
nums.append(40)
nums[0] = 99
```

```python
nums
```

## Dictionary
- in Python, a dictionary (or dict) is a built-in data structure used to store data in key-value pairs. It is a "mapping" type where each unique key is associated with a specific value, much like a real-world dictionary maps a word to its definition

```python
data = {
    "name": "Ali",
    "age": 20
}

print(data["name"])
```

## tuple
- In Python, a tuple is a built-in collection data type used to store multiple items in a single variable. It is a sequence of objects that is ordered, immutable, and allows duplicate values.

```python
t = (10, 20, 30)
```

```python
t
```

## Set
- In Python, a set is a built-in data type used to store a collection of items that is unordered, unique, and mutable. Unlike lists or tuples, a set cannot contain duplicate elements; if a duplicate is added, it is automatically ignored.

```python
s = {1, 2, 3, 3}
print(s)   
```

```python
a = {1,2,3}
b = {3,4,5}

print(a & b)  # intersection
print(a | b)  # union
```
