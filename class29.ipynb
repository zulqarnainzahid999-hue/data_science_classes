Base Class (ElectronicDevice): This is the parent class: Attribute: name, price, power_status
Methods: (power_on()) that all electronic devices share.

Subclass (Laptop): This is the child class. It inherits (automatically receives) all the attributes and methods from ElectronicDevice. Attributes:(screen_size_inches, ports) Methods: list_ports()

```python
class electronic_device:

    def __init__(self, name, price, power_status):
        self.name = name
        self.price = price
        self.power_status = power_status

    def power_on(self):
        print('power is on')
```

```python
class laptop(electronic_device):
    def __init__(self, name, price, power_status, screen_size, ports):
        super().__init__(name, price, power_status)
        self.screen_size = screen_size
        self.ports = ports

    def list_ports(self):
        print(f'these are the ports: {', '.join(self.ports)}')
```

```python
l1 = laptop('HP', 25000, 'runing', '14 inch', ['usb type_a', 'RJ_45', 'card reader'])
```

```python
l1.ports
```

```python
l1.list_ports()
```

```python
class User:
    def __init__(self, fname, lname, age):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.fullname = f"{self.fname} {self.lname}"

    def describe_user(self):
        print(f"Introduction\n{self.fullname}\n{self.age}")

    def greet_user(self):
        print(f"Hello {self.fname}, welcome to our page")
```

```python
user1 = User('Zulqarnain', 'Zahid', 16)
```

```python
user1.describe_user()
```

```python
class admin(User):
    def __init__(self, fname, lname, age, privilages):
        super().__init__(fname, lname, age)
        self.privilages = privilages

    def show_privalages(self):
        for privilage in self.privilages:
            print(f'_{privilage}')
```

```python
zulqarnain = admin('Zulqarnain', 'Zahid', 16, ['coding', 'teaching'])
```

```python
class Car:
    def __init__(self,make, model, year):
        # Attributes
        self.make = make
        self.model = model
        self.year =  year 
    
    # Methods
    def apply_accelerator(self):
        print('Accelerator is applied')
    
    def apply_breaks(self):
        print('Brakes are applied')
    
```

```python
class Battery:
    def __init__(self, company, voltage, capacity):
        self.company = company  
        self.voltage = voltage  
        self.capacity = capacity  

    def battery_info(self):
        print(f'Company: {self. company}\nVoltage:{self.voltage}\nCapacity: {self.capacity}')
```

```python
class ElectricCar(Car):
    def __init__(self,make, model, year, company, voltage, capacity):
        super().__init__(make,model,year)

        # Additional Attribute
        self.battery = Battery(company, voltage, capacity)


    # Additional Method
    def recharge(self):
        print('The battery is recharged now ')
```

```python
newec =ElectricCar('tesla', 'cybertruck', 2021, 'osaka', '12v', '50ah')
```

```python
newec.battery.battery_info()
newec.battery.company
```

```python
newec.apply_accelerator()
```
