# Merge

```python
import pandas as pd

df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Ali", "Sara", "Ahmed"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3],
    "marks": [80, 90, 85]
})

df = pd.merge(df1, df2, on="id")
print(df)
```

# groupby

```python
import pandas as pd

data = pd.DataFrame({
    "department": ["IT", "HR", "IT", "HR"],
    "salary": [50000, 40000, 60000, 45000]
})

result = data.groupby("department")["salary"].mean()
print(result)
```
