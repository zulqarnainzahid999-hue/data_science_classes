## Arbitrary num of arguments

```python
def any_function(param1):
    print(param1)
```

```python
any_function('Zulqarnain Zahid')
```

```python
def make_pizza(*Toppings):
    print(Toppings)
```

```python
make_pizza('peproni', 'onion')
make_pizza('peproni', 'onion', 'bacon')
make_pizza('peproni', 'onion', 'bacon', 'prosciuto')
```

```python
def make_pizza(size, *Toppings):
    print(f'making {size}_inch pizza using following Toppings:')
    for Topping in Toppings:
        print(f' - {Toppings}')
```

```python
make_pizza('15', 'peproni, onion, bacon')
```

```python
make_pizza('12', 'paeproni')
```

```python
def build_profile(**info):
    print(info)
```

```python
build_profile(Fname= 'Zulqarnain', Lname='Zahid')
build_profile(City= 'karachi', Course= 'python')
```

```python
def make_sandwich(*items):
    print(f'this sandwich include the following items')
    for  item in items:
        print(f'- {items}')
```

```python
make_sandwich('bread, tomato, onion')
make_sandwich('bread, onion, tomato, chili')
```

```python
def make_car(manufacturer, model, **info):
    if (info):
        print(f'this car is manufactured by {manufacturer} and its model is {model}:')
        for key, value in info.items():
            print(f'{key} : {value}') 
    else:
        print(f'this car is manufactured by {manufacturer} and its model is {model}')
    
```

```python
make_car('honda', 'civic')
make_car('honda', 'civic', engine= '1200CC', AC= 'Runing')
```

```python
make_car('BMW', 'M5', Engine= 'V8', Color= 'Black')
```
