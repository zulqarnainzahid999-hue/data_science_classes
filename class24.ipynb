```python
with open('learning_python.txt', 'w+') as f:
    f.write('I love python\n')
    f.write('I practice python\n')
    f.write('I sometimes stuck in python')

    f.seek(0)
    content = f.read()
    new_content = content.replace('python', 'java')

    f.seek(0)
    f.write(new_content)
    f.truncate()
    print(new_content)
```
