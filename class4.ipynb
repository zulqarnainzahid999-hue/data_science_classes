```python
# Snake Game using Tkinter
# Save as snake.py and run: python snake.py

import tkinter as tk
import random

# ----- Game configuration -----
CELL_SIZE = 20      # size of one grid cell (pixels)
GRID_WIDTH = 30     # number of cells horizontally
GRID_HEIGHT = 20    # number of cells vertically
DELAY = 100         # movement delay in ms (smaller -> faster)

# ----- Game logic -----
class SnakeGame:
    def __init__(self, root):
        self.root = root
        self.width_px = CELL_SIZE * GRID_WIDTH
        self.height_px = CELL_SIZE * GRID_HEIGHT

        self.canvas = tk.Canvas(root, width=self.width_px, height=self.height_px, bg="#0b3d91")
        self.canvas.pack()

        self.score = 0
        self.direction = "Right"   # initial direction
        self.running = False

        # UI: score text and messages
        self.score_text = self.canvas.create_text(10, 10, anchor="nw",
                                                  fill="white", font=("Arial", 14, "bold"),
                                                  text=f"Score: {self.score}")
        self.message = self.canvas.create_text(self.width_px//2, self.height_px//2,
                                               fill="white", font=("Arial", 20, "bold"))

        # Bind keys
        root.bind("<Up>", lambda e: self.change_direction("Up"))
        root.bind("<Down>", lambda e: self.change_direction("Down"))
        root.bind("<Left>", lambda e: self.change_direction("Left"))
        root.bind("<Right>", lambda e: self.change_direction("Right"))
        root.bind("r", lambda e: self.restart())  # press 'r' to restart

        self.restart()

    def restart(self):
        # reset game state
        self.canvas.delete("all")
        # recreate HUD
        self.score = 0
        self.score_text = self.canvas.create_text(10, 10, anchor="nw",
                                                  fill="white", font=("Arial", 14, "bold"),
                                                  text=f"Score: {self.score}")
        self.message = self.canvas.create_text(self.width_px//2, self.height_px//2,
                                               fill="white", font=("Arial", 20, "bold"),
                                               text="Press Arrow Key to Start")
        # snake: list of (x,y) cells. start at center
        start_x = GRID_WIDTH // 2
        start_y = GRID_HEIGHT // 2
        self.snake = [(start_x - i, start_y) for i in range(3)]  # length 3, horizontal
        self.direction = "Right"
        self.place_food()
        self.running = False

        # draw initial snake + food
        self.draw_everything()

    def place_food(self):
        # choose a random cell not occupied by snake
        empty_cells = [(x, y) for x in range(GRID_WIDTH) for y in range(GRID_HEIGHT) if (x, y) not in self.snake]
        self.food = random.choice(empty_cells)

    def change_direction(self, new_dir):
        # start running on first key press
        if not self.running:
            self.canvas.itemconfig(self.message, text="")

        # prevent reversing directly
        opposites = {"Up":"Down", "Down":"Up", "Left":"Right", "Right":"Left"}
        if opposites.get(new_dir) == self.direction:
            return
        self.direction = new_dir
        if not self.running:
            self.running = True
            self.game_loop()

    def game_loop(self):
        if not self.running:
            return

        head_x, head_y = self.snake[0]
        if self.direction == "Up":
            head_y -= 1
        elif self.direction == "Down":
            head_y += 1
        elif self.direction == "Left":
            head_x -= 1
        elif self.direction == "Right":
            head_x += 1

        new_head = (head_x, head_y)

        # check collisions with walls
        if (head_x < 0 or head_x >= GRID_WIDTH or head_y < 0 or head_y >= GRID_HEIGHT):
            self.game_over()
            return

        # check collision with itself
        if new_head in self.snake:
            self.game_over()
            return

        # move snake
        self.snake.insert(0, new_head)

        # check food
        if new_head == self.food:
            self.score += 1
            # increase speed slightly every few points
            global DELAY
            if self.score % 5 == 0 and DELAY > 30:
                DELAY -= 5
            self.place_food()
        else:
            # remove tail
            self.snake.pop()

        self.draw_everything()
        self.root.after(DELAY, self.game_loop)

    def draw_everything(self):
        self.canvas.delete("cell")
        # draw food
        fx, fy = self.food
        self.draw_cell(fx, fy, fill="#ffcc00", tag="cell")
        # draw snake
        for i, (x, y) in enumerate(self.snake):
            color = "#00ff88" if i == 0 else "#00cc66"  # head different color
            self.draw_cell(x, y, fill=color, tag="cell")
        # update score
        self.canvas.itemconfig(self.score_text, text=f"Score: {self.score}")

    def draw_cell(self, gx, gy, fill="#ffffff", tag=""):
        x1 = gx * CELL_SIZE + 2
        y1 = gy * CELL_SIZE + 2
        x2 = x1 + CELL_SIZE - 4
        y2 = y1 + CELL_SIZE - 4
        # rounded-ish rectangles using create_rectangle (simple)
        self.canvas.create_rectangle(x1, y1, x2, y2, fill=fill, outline="#063d2f", width=2, tags=tag)

    def game_over(self):
        self.running = False
        self.canvas.itemconfig(self.message, text=f"Game Over! Score: {self.score}\nPress 'r' to restart")
        # optionally draw a semi-transparent overlay
        self.canvas.create_rectangle(0, 0, self.width_px, self.height_px, fill="#000000", stipple="gray50", tags="overlay")
        self.canvas.tag_raise(self.message)

# ----- Run the game -----
if __name__ == "__main__":
    root = tk.Tk()
    root.title("Snake Game")
    # prevent window from being resized (keeps grid consistent)
    root.resizable(False, False)

    game = SnakeGame(root)
    root.mainloop()
```
