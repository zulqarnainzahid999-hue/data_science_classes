```python
def sports_detail():
    flag = True
    data = {}
    while flag:
        name = input('enter name').title()
        sport = input('enter sport').title()
        if name == 'Q' or sport == 'Q':
            flag = False
        else:
            if name not in data:
                data[name] = [sport]
            else:
                data[name].append(sport)
    with open('testing1.txt', 'a') as f:
         for name,sport in data.items():
             f.write(f'{name}: {','.join(sport)}\n')            
```

```python
sports_detail()
```
