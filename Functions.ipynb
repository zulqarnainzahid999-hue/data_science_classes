```python
def get_even_odd_numbers(numbers):
    even = []
    odd = []
    for num in numbers:
        if num % 2 == 0:
            even.append(num)
        else:
            odd.append(num)
    print(even , odd , sep= '\n')
```

```python
get_even_odd_numbers([70, 5, 74,9])
```

```python
def get_even_odd(numbers):
    even = []
    odd = []
    for num in numbers:
        if num % 2 == 0:
            even.append(num)
        else:
            odd.append(num)
    print(even, odd, sep='\n')
```
