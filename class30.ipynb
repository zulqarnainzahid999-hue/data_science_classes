## Polymorphism

The one way of doing polymorphism is redefining the parent class method with child class
Method overloading

Method overloading is like having a function that can do different things based on the number or type of parameters it receives. It allows a class to define multiple methods with the same name but different parameter signatures.
Python does not support traditional method overloading like some other languages.

```python
def add(a, b):
    print(a + b)
```

```python
def add(e, d, f):
    print(e + d + f)
```

```python
add(1, 2, 3)
```

```python
# operator overloading
print(2 + 3)
print('Zulqarnain' + 'Zahid')
```

Method overriding
- same methods in different class

```python
class electriccar(car):
    def __init__(self, brand, model, transmission, engine, battery):
        super().__init__(brand, model, transmission, engine)
        self.battery = battery

    def recharge_battery(self):
        print('battery is recharged')

    def fuel_tank(self, car):
        if car:
            print(f'{self.car_brand} has no fuel tank')
        else:
            super().fill_fuel()
```

```python
ec1 = electriccar('Tesla', '22', 'Auto', '600cc', 'Osaka')
```

```python
ec1.fill_fuel()
```

## Encapulsation
- The method of binding attributes and methods together in a class.

```python
class ElectricCar(car):
    def __init__(self, brand, model, transmission, engine, battery):
        self.__battery = battery
        super().__init__(brand, model, transmission, engine)
        
    def chargebattery(self):
        print('Batter is being charged')
        
    def fueltank(self):
        print(f'{self.car_brand} has no fuel tank')
    
    def getbattery(self):
        print(f'{self.car_brand} has {self.__battery} battery installed')
        
    def setbattery(self, newname):
        self.__battery = newname
```

```python
ec1 = ElectricCar('Ford', '22', 'Auto', '700cc', 'Atlas')
```

```python
ec1.setbattery('osaka')
```

# Multiple Inheritence (without constructor)

```python
class a():
    def feature1(self):
        print('festure 1 from class a is working')
    def feature2(self):
        print('feature 2 from claa a is working')
```

```python
class b():
    def feature3(self):
        print('feature 3 from class b is working')

    def feature4(self):
        print('feature 4 from class b is working')
```

```python
class c(a,b):
    def feature5(self):
        print ('feature 5 from class c is woring')
```

```python
C = c()

C.feature1()
C.feature2()
C.feature3()
C.feature4()
C.feature5()
```

- Key notes:

- In multilevel inheritence, the class C having all the features of both class A and B, but for having the features of class A, class C is dependent on class B therefore, if we make any changes in the methods of class A in class B (polymorphism), they are also effected which means class C can not directly access the features of class A without being dependent on class B, to resolve this issue, multiple inheritence comes into place

- In multiple inheritence, all the features of class A and class B are present in class C, but Class C is not dependent on class B to having the features of class A, Class C is directly inheriting the features of both class A and class C

## Multiple Inheritence (with __ init __)

```python
class A():
    
    def __init__(self):
        print('In init of A')
    
    def feature1(self):
        print('Feature 1 from class A is working')
    def feature2(self):
        print('Feature 2 from class A is working')
        
    def show(self):
        print('I am showing class A')
 
```

```python
class B():
    
    def __init__(self):
        print('In init of B')
            
    def feature3(self):
        print('Feature 3 from class B is working')
    def feature4(self):
        print('Feature 4 from class B is working')
        
    def show(self):
        print('I am showing class B')
```

```python
class C(A,B):
    
    def __init__(self):
        print('In init of C')
        super().__init__()
        
        # we have two parent classes super will call init of ??
        # There is a term called MRO (method resolution order)
        # MRO is from left to right
        # init of A will be called (it also applies on methods)
```

```python
c = C()
```

```python
class C(B,A):
    
    def __init__(self):
        print('In init of C')
        super().__init__()
        
        # we have two parent classes super will call init of ??
        # There is a term called MRO 
        # MRO is from left to right
        # init of A will be called (it also applies on methods)
```

```python
c = C()
```

```python
class A():
    
    def __init__(self):
        print('In init of A')
    
    def feature1(self):
        print('Feature 1 from class A is working')
    def feature2(self):
        print('Feature 2 from class A is working')
        
    def show(self):
        print('I am showing class A')
 
class B():
    
    def __init__(self):
        print('In init of B')
            
    def feature2(self):
        print('Feature 2 from class B is working')
    def feature4(self):
        print('Feature 4 from class B is working')
        
    def show(self):
        print('I am showing class B')
        

class C(A,B):
    
    def __init__(self):
        print('In init of C')
        super().__init__()
        super().show()
        
    def feature2(self):
        print('I am in C') 
    def feat(self):
        super().feature2()     
```

```python
c = C()
```

```python
c.feat()
```

## Abstraction
Abstract Class and Methods in Python

- Python does not support Abstraction (since python is a dynamically typed language and we do not have any concern with the data type, the concept of abstraction applies on data type)
- We will use a module ABC for abstraction
- ABC means Abstract Base Classes

```python
# A normal class and a normal method
class Computer:
    def process(self):
        print('running')
```

```python
# A method that only has declaraion but has nothing in it is abstract method

class Computer():
    def process(self):
        pass

# A class having a method that has no body
```

- Hiding the implementation details of a method is called Abstraction
- We can not create an object of abstract class

```python
com1 = Computer()
com1.process()

# There is no error and we are able to create object and call method since its not
#  an abstract class and not an abstract method
```

```python
from abc import ABC, abstractmethod
```

```python
class Computer(ABC):
    @abstractmethod  #decorator
    def process(self):  # abstract method
        pass
    
    @abstractmethod
    def greet(self):
        pass
        
        # to make a class abstract 
        
        # it must inherit the ABC class from abc module
        # it must have atleast one abstract method(which is defined using
        #  a decorator @abstractmethod)
```

```python
com = Computer()
# We can not create objects of abstract class
```

```python
class Laptop(Computer):
    def process(self):
        print('It is running')
        
lap = Laptop()
```

```python
class Laptop(Computer):
    def process(self):
        print('It is running')
        
    def greet(self):
        print('Salam')
        
lap = Laptop()
```

```python
lap.greet()
```

### What is the use of this concept or functionality
- Through abstraction we can provide a user an interface that only show the behaviour not the implementaion of that behaviour, like we show a suer a computer that runs process and user can only see the name of behaviour but there is no implementation in it.

- Implementation is done in child class

```python
def factorial():
    num = int(input('enter your number'))
    result = 1
    for i in range(num,0, -1):
        result = i * result
    return result
```
