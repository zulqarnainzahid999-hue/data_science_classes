```python
add = lambda a, b: a + b
```

```python
print(add(3, 3))
```

```python
add3 = lambda a, b, c: a / b * c
```

```python
add3(5, 5, 5)
```

```python
nums = [(1, 5), (2, 3), (4, 1)]
nums.sort(key=lambda x: x[1])
```

```python
nums
```

```python
nums = [1, 2, 3, 4, 5]

even = list(filter(lambda x: x % 2 == 0, nums))
print(even)   # [2, 4]
```

```python
nums = [1, 2, 3]

double = list(map(lambda x: x * 2, nums))
print(double)   # [2, 4, 6]
```
