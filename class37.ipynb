```python
import pandas as pd
def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    customer_name = []
    for ids, name in zip(customers['id'], customers['name']):
        if ids not in orders['customerids'].values:
            custome_name.append(name)
    return pd.Dataframe({'customers': customer_name})
```

```python
import pandas as pd

def article_views(views: pd.DataFrame) -> pd.DataFrame:
    views = views.query('author_id == viewer_id')
    views.drop_duplicates(subset=['author_id','viewer_id' ], inplace=True)
    result = views.sort_values(by='author_id')[['author_id']]
    return result.rename(columns={'author_id':'id'})
    # print(views['author_id'] == views['viewer_id'])
    
    # print(result)
```
