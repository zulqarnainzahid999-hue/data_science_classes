```python
def make_album():
    flag = True
    result = {}
    while flag:
        artist = input('Enter artist name: ')
        album = input('Enter album name: ')
        result[artist] = album
        if artist.lower:
            choice = input('Do you want to continue? yes/no')
            if choice == 'no':
                flag = False
                break
      
    return result
    
albums = make_album()
print(albums)
```
