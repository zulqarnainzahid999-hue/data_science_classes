```python
from scipy.stats import chisquare

observed = [20, 30, 50]
expected = [33, 33, 34]

chisquare(observed, expected)
```

```python
from scipy.stats import f_oneway

group1 = [10,20,30]
group2 = [15,25,35]
group3 = [20,30,40]

f_oneway(group1, group2, group3)
```
